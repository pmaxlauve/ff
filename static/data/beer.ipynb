```python
import pandas as pd
```

```python
draft_df = pd.read_csv('beersheet.csv')
draft_df
```

```python
draft_df = draft_df.dropna(subset = ['ECR'])
draft_df
```

```python
team = []
bye = []

for x in range(len(draft_df)):
    tb = draft_df.iloc[x, 2].split('/')
    
    team.append(tb[0])
    
    bye.append(tb[1])

draft_df['Team'] = team 
draft_df['Bye'] = bye
draft_df
```

```python
df = draft_df[['Name', 'Pos', 'Rank', 'Tier', 'ECR', 'Team', 'Bye']]
df
```

```python
n=0
pid = []

for x in range(len(df)):
    n+=1
    pid.append(n)

    
df['ID'] = pid
df
    
```

```python
df.columns = ['name', 'pos', 'rank', 'tier', 'ecr', 'team', 'bye', 'id']
df
```

```python
df['ecr2'] = [x.split('|') for x in df['ecr']]

df
```

```python
r = []
s = []
for x in range(len(df)):
    rnd = int(df.iloc[x, 8][0])
    spt = int(df.iloc[x, 8][1])
    
    r.append(rnd)
    s.append(spt)

    
df['round'] = r
df['spot'] = s
df =df.sort_values(by=['round', 'spot'])
df
```

```python
df = df[['name', 'pos', 'rank', 'tier', 'ecr', 'team', 'bye', 'id']]
df
```

```python
df.to_csv('beer.csv', index = False)
```
